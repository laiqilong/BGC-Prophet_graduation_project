```python
with open("easy_MIBIG_cdhit_cluster.tsv", "r") as f:
    text = f.readlines()
    clusters = {}
    rep_clusters = {}
    for line in text:
        represent, member = line.split()
#         print(represent, member)
        if represent in clusters:
            clusters[represent].add(member)
        else:
            clusters[represent] = {represent,}
        
        rep_clusters[member] = represent
print(len(rep_clusters))
print(len(clusters))
print(clusters)
```

```python
import re
```

```python
with open("mibig_cdhit.clstr", "r")as f:
    text = f.readlines()
    text = ''.join(text)
    pattern = re.compile(r">Cluster")
    cluster_start = [substr.start() for substr in re.finditer(pattern, text)] + [len(text)-1]
    for i in range(len(cluster_start)-1):
        block = text[cluster_start[i]:cluster_start[i+1]-1]
#         print(block)
        block_lines = block.split('\n')
        mems = set()
        rep = None
        for line in block_lines:
            if "BGC" in line:
                startN = line.find("BGC")
                endN = line.find(".")
                # -1
                BGCs = line[startN:endN]
                if " " in BGCs:
                    print("")
                    break
                mems.add(BGCs)
                if not rep and "*" in line:
                    rep = BGCs
                    print(line)
        if not rep:
            print("")
            break
            mmseqs_rep = rep_clusters[rep]
            clusters[mmseqs_rep]  = clusters[mmseqs_rep] | mems
            
#         print(len(block_lines))
        
```
