```python
folderPath = 'mibig_gbk_3.1/'
def get_label(file, labels):
    with open(folderPath+file, 'r') as fp:
        text = fp.readlines()
        for line in text:
            if r'/label=' in line:
                label = line[line.find('=')+2:-2]
                labels[file.split('.')[0]] = label
                break

labels = {}
for file in os.listdir(folderPath):
    get_label(file, labels)
```

```python
for k, v in labels.items():
    if len(v)>1:
        tmp = [(it, labels_count[it]) for it in v]
        tmp.sort(key=lambda a:a[1])
        suitable_v = tmp[0][0] if tmp[0][1]>1 else tmp[1][0]
    else:
        suitable_v = v[0]
    labels[k] = suitable_v
```
